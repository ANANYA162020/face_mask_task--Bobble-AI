<h2> Code Here </h2>

```
def crop_by_mask(image, mask):
    ''' Your code here '''
    img_mask = cv2.inRange(mask, (0,0,0), (0,0,0))
    result_image = np.bitwise_and(image,mask,image) # instead of np.bitwise_and(image,mask,image) we could have also used np.cv2
    result_image = cv2.bitwise_and(image,mask,image)
    return result_image


    
```
